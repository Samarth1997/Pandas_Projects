Create column for Time stamps


```python
df.loc[3].Title
def Time_stmp(txt1):
    check=False
    data=''
    for i in range (len(txt1)):
        if txt1[i]==")":
            for j in range(i+1,i+20):
                data+=txt1[j]
            return data
  
df['Time_stamp']=df['Title'].apply(Time_stmp)
df.head()
# df['Time_stamp'].drop
```

Retriving Number of months from time stamp column

```python
from dateutil import parser

def calculate_months(time_range):

    try:
        start_str, end_str = time_range.split(' - ')
        start_date = parser.parse(start_str)
        end_date = parser.parse(end_str)
        total_months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) + 1
        return total_months
    except Exception as e:
        print(f"Error parsing '{time_range}': {e}")
        return None

# Apply function to the column
df['Total_months'] = df['Time_stamp'].apply(calculate_months)

# Display result
df.head()
```
